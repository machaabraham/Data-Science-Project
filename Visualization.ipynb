```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import math
dataset=pd.read_csv("Infant-Mortality-Rate-Tz.csv")
#Let find the avarege/mean infant mortality rate below
import numpy as np
np_mortality_rate=np.array(dataset.NUMERIC)
np_mortality_year=np.array(dataset.YEAR)
avarage_mortality_rate=np.mean(np_mortality_rate)
print(avarage_mortality_rate)
print(np.min(np_mortality_year))
np.corrcoef(np_mortality_rate,np_mortality_year)
```

```python
plt.scatter(dataset.YEAR,dataset.NUMERIC,color='red')
plt.title("Graph of Year Vs Infant Mortality rate")
plt.ylabel("YEAR")
plt.xlabel("Mortality Rate")
plt.xlim(np.min(dataset.YEAR),np.max(dataset.YEAR))
plt.ylim(np.min(dataset.NUMERIC),np.max(dataset.NUMERIC))
plt.legend()
plt.show()
```

```python
plt.bar(dataset.YEAR,dataset.NUMERIC,color='green',width=0.5)
plt.xlabel("YEAR")
plt.ylabel("Mortality Rate")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(dataset.NUMERIC, labels=dataset.YEAR)
plt.legend()
plt.show()
```
